```python
from astropy import units as u
from astropy import constants as c
import math
import matplotlib.pyplot as plt
%inline
```

```python
def R(E, V):
    m = c.m_e.value
    h = c.hbar.value
    E = ((E * u.eV).to(u.J)).value
    V = ((V * u.eV).to(u.J)).value
    k1 = math.sqrt(2*m*E/h**2)
    k2 = math.sqrt(2*m*(E-V)/(h**2))
    R = abs((k1-k2)/(k1+k2))**2
    return R
```

```python
E0 = []
R0 = []
for i in range(100, 300):
    E0 += [i/10]
    R0 += [R(i/10,10)]
T0 = [1-i for i in R0]
```

```python
plt.plot(E0, R0, E0, T0)
```
